```python
# special tokens for prompting
system_token = "<SYSTEM_TASK:>"
user_token = "<USER_TASK:>"
assistant_token = "<ASSISTANT_TASK:>"
end_token = "<END_TASK>"
```

```python
def process_1(data):
  start = 'Given the following text description, write Python code to implement the functionality described.'
  des = 'Description:'
  code_prompt = 'Python Code:'
  text_prompts = list(map(lambda x: x.replace(' | ', '\n'), data['text']))
  prompt = ''
  final_code = ''
  final_text_prompt = []
  final_code_prompt = []
  code = data['code']
  for i in range(len(text_prompts)):
    prompt = system_token + '\n' + start + '\n' + end_token + '\n' + user_token + '\n' +  des + '\n' + text_prompts[i] + '\n'
    code[i] = code[i].replace('NEW_LINE INDENT' , '\n\t')
    code[i] = code[i].replace('NEW_LINE' , '\n')
    code[i] = code[i].replace('DEDENT' , '\b')
    code[i] = code[i].replace(' ( ' , '(')
    code[i] = code[i].replace(' [ ', '[')
    code[i] = code[i].replace('_ ', '_')
    code[i] = code[i].replace('" ', '"')
    final_code = assistant_token + '\n' + code_prompt + '\n' + code[i] + '\n' + end_token + '\n'
    final_text_prompt.append(prompt)
    final_code_prompt.append(final_code)
  return {
      'text_prompt' : final_text_prompt,
      'code_prompt' : final_code_prompt
  }
```

```python
import re
from functools import reduce
import os

def process_3(data):
  start = 'Given the following text description, write Python code to implement the functionality described.'
  des = 'Description:'
  code_prompt = 'Python Code:'
  final_prompt = ''
  final_code = ''
  final_text_prompt = []
  final_code_prompt = []

  # extract and delete text prompt then combine canonical solution and prompt
  unclean_prompt = data['prompt']
  occurrences = re.finditer('"""', unclean_prompt)
  # using reduce() to get start indices of all occurrences
  res = reduce(lambda x, y: x + [y.start()], occurrences, [])
  extracted_prompt = ''
  if len(res)%2 == 0:
    for i in range(int(len(res)/2)):
      prompt = unclean_prompt[res[2*i] + 3 : res[2*i + 1]]
      extracted_prompt = extracted_prompt + prompt + '\n'
      unclean_prompt = unclean_prompt.replace(prompt , '')
      unclean_prompt = unclean_prompt.replace('"""' , '')
      extracted_prompt = extracted_prompt.replace('>>>' , '\nThis is how the function will work:\n')
    code = unclean_prompt + data['canonical_solution']
    for i in range(len(data['prompt'])):
      final_prompt_unclean = system_token + '\n' + start + '\n' +  end_token + '\n' + user_token + '\n' + des + '\n' + extracted_prompt + '\n'
    final_code_unclean = assistant_token + '\n' + code_prompt + '\n' + code + '\n' + end_token + '\n'
    # removing empty lines from code and prompt
    final_code = os.linesep.join([line for line in final_code_unclean.splitlines() if line])
    final_prompt = os.linesep.join([line for line in final_prompt_unclean.splitlines() if line])
  else:
    print('Error :( ')
  return {
      'text_prompt' : final_prompt,
      'code_prompt' : final_code
  }
```

```python
import re
from functools import reduce
import os

def process_5(data):
  start = 'Given the following text description, write Python code to implement the functionality described below step by step'
  des = 'Description:'
  code_prompt = 'Python Code:'
  final_prompt = ''
  final_code = ''
  final_text_prompt = []
  final_code_prompt = []
  # extract and delete text prompt then combine canonical solution and prompt
  unclean_prompt = data['content']
  occurrences = re.finditer('"""', unclean_prompt)
  # using reduce() to get start indices of all occurrences
  res = reduce(lambda x, y: x + [y.start()], occurrences, [])
  extracted_prompt = ''
  if len(res)%2 == 0:
    for i in range(int(len(res)/2)):
      if i == 0:
        raw_prompt = unclean_prompt[res[2*i] + 3 : res[2*i + 1]]
        if '\nExplanation' in raw_prompt:
          arr = raw_prompt.split(':')
          if '\nExplanation' in  arr:
            i1 = arr.index('\nExplanation')
            extracted_prompt = extracted_prompt + arr[i1+ 1] + '\n'
      else:
        raw_prompt = unclean_prompt[res[2*i] + 3 : res[2*i + 1]]
        if '\nExplanation' in raw_prompt:
          arr = raw_prompt.split(':')
          if '\nExplanation' in  arr:
            i1 = arr.index('\nExplanation')
            extracted_prompt = extracted_prompt + f'Step{i}: ' + arr[i1+1] + '\n'
            if '\nEnd of explanation' in extracted_prompt:
              extracted_prompt = extracted_prompt.replace('\nEnd of explanation', '')
      code = unclean_prompt.replace(extracted_prompt , '')
      code = unclean_prompt.replace('"""' , '')
    final_prompt_unclean = system_token + '\n' + start + '\n' +  end_token + '\n' + user_token + '\n' + des + '\n' + extracted_prompt + '\n'
    final_code_unclean = assistant_token + '\n' + code_prompt + '\n' + code + '\n' + end_token + '\n'
    # removing empty lines from code and prompt
    final_code = os.linesep.join([line for line in final_code_unclean.splitlines() if line])
    final_prompt = os.linesep.join([line for line in final_prompt_unclean.splitlines() if line])
  else:
    print('Error :(')
  return {
      'text_prompt' : final_prompt,
      'code_prompt' : final_code
  }
```

```python
import re
from functools import reduce
import os

def process_4(data):
  start = 'Given the following text problem statement, write Python code to implement the functionality described below in problem statement'
  code_prompt = 'Python Code:'
  des = 'Problem:'
  final_prompt = ''
  final_code = ''
  final_text_prompt = []
  final_code_prompt = []
  unclean_prompt = data['prompt']
  extracted_prompt = unclean_prompt.split('\n')
  for i in range(1,len(extracted_prompt)):
     if extracted_prompt[i] != '':
      prompt = extracted_prompt[i]
      break
  occurrences_start = re.finditer('<code>', unclean_prompt)
  # using reduce() to get start indices of all occurrences
  res_start = reduce(lambda x, y: x + [y.start()], occurrences_start, [])
  occurrences_end = re.finditer('</code>', unclean_prompt)
  # using reduce() to get start indices of all occurrences
  res_end = reduce(lambda x1, y1: x1 + [y1.start()], occurrences_end, [])
  if len(res_start) != 0  and len(res_end) != 0:
    code_extracted = unclean_prompt[res_start[0] + 6 : res_end[0]]
    code = data['code']
    final_code_unclean =    assistant_token + '\n' + code_prompt + '\n' + code_extracted + '\n' + code + '\n' + end_token + '\n'
    final_prompt_unclean = system_token + '\n' + start + '\n' +  end_token + '\n' + user_token + '\n' + des + '\n' + prompt + '\n'
    # removing empty lines from code and prompt
    final_code = os.linesep.join([line for line in final_code_unclean.splitlines() if line])
    final_prompt = os.linesep.join([line for line in final_prompt_unclean.splitlines() if line])
  return {
      'text_prompt' : final_prompt,
      'code_prompt' : final_code
  }
```
